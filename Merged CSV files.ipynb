```python
# Dependencies and Setup
import pandas as pd

# Study data files
GDP_data_path = "./GDP.csv"
stock_data_path = "./Stock.csv"

# Read the data 
GDP_data = pd.read_csv(GDP_data_path)
stock_data = pd.read_csv(stock_data_path)
```

```python
GDP_clean_df = GDP_data.drop(['Unnamed: 0', 'realtime_start', 'realtime_end'], axis=1)
GDP_clean_df['date'] = pd.to_datetime(GDP_clean_df['date'])
GDP_clean_df = GDP_clean_df.set_index('date')
GDP_clean_df = GDP_clean_df.rename(columns={"value":"GDP"})
GDP_clean_df.drop(index=GDP_clean_df[GDP_clean_df['GDP'] == '.'].index, inplace=True)
GDP_clean_df['GDP'] = GDP_clean_df['GDP'].astype(float)
GDP_clean_df.tail(30)
```

```python
stock_clean_df = stock_data.drop(['Unnamed: 0','Microsoft Open','Apple Open', 'Amazon Open', 'IBM Open', 'Nike Open'], axis=1)
stock_clean_df['Date'] = pd.to_datetime(stock_clean_df['Date'])
stock_clean_df = stock_clean_df.rename(columns={"Date":"date", "Close Microsoft":"Microsoft", "Apple Close":"Apple", "Amazon Close":"Amazon", "IBM Close":"IBM", "Nike Close":"Nike"})
stock_clean_df = stock_clean_df.set_index('date')
stock_clean_df.head()
```

```python
# Merge data into single dataset
data_df = pd.merge(GDP_clean_df, stock_clean_df, on='date')

# Display the data table for preview
data_df.head()
```

```python
# Create a csv file
data_df.to_csv('Dataset.csv')
stock_clean_df.to_csv('Stocks_clean.csv')
GDP_clean_df.to_csv('GDP_Clean.csv')
```
