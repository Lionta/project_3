```python
# Dependencies and Setup
import pandas as pd

# Study data files
GDP_data_path = "./GDP.csv"
stock_data_path = "./Stock.csv"

# Read the data 
GDP_data = pd.read_csv(GDP_data_path)
stock_data = pd.read_csv(stock_data_path)
```

```python
# Merge data into single dataset
data_df = pd.merge(GDP_data, stock_data)

# Display the data table for preview
data_df.head()
```

```python
# Create a csv file
data_df.to_csv('Dataset.csv')
```
